```python
# Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the Dataset

```python
# Load the dataset

df = pd.read_csv("../data/superstore.csv", encoding="latin1")

df.head()
```

## Check for Duplicate Records

```python
# Check for duplicate rows

duplicates = df.duplicated().sum()

print("Number of duplicate rows:", duplicates)
```

## Data Preprocessing
### Convert Date Columns to Datetime

```python
# Convert date columns to datetime format

df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

# Check the updated data types
df.info()
```

## Descriptive Statistics

```python
# Display descriptive statistics

df.describe()
```

## Sales Distribution

```python
# Create a histogram for Sales

plt.figure(figsize=(10, 6))


sns.histplot(
    df["Sales"],
    bins=30,
    kde=True
)

plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Number of Orders")

plt.grid(alpha=0.3)

plt.savefig("../images/sales_distribution.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

### Business Insights

- Most orders have relatively low sales values.
- Only a small number of orders have very high sales values.
- The sales distribution is positively (right) skewed.
- A few high-value orders act as outliers and stretch the distribution.
