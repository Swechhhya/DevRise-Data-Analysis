## Load the Dataset

```python
# Load the dataset

df = pd.read_csv("../data/superstore.csv", encoding="latin1")

df.head()
```

## Data Preprocessing
### Convert Date Columns to Datetime

```python
# Convert date columns to datetime format

df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

# Check the updated data types
df.info()
```

## Descriptive Statistics

```python
# Display descriptive statistics

df.describe()
```

## Sales Distribution

```python
# Create a histogram for Sales

plt.figure(figsize=(10, 6))


sns.histplot(
    df["Sales"],
    bins=30,
    kde=True
)

plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Number of Orders")

plt.grid(alpha=0.3)

plt.savefig("../images/sales_distribution.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

### Business Insights

- Most orders have relatively low sales values.
- Only a small number of orders have very high sales values.
- The sales distribution is positively (right) skewed.
- A few high-value orders act as outliers and stretch the distribution.

Sales by Category

```python
# Group sales by category
category_sales = df.groupby("Category")["Sales"].sum().reset_index()

# Plot
plt.figure(figsize=(8,5))

sns.barplot(
    x="Category",
    y="Sales",
    data=category_sales
)

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.grid(alpha=0.3)

plt.savefig("../images/sales_by_category.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Business Insights

- Technology category generates the highest total sales.
- Furniture and Office Supplies generate lower sales compared to Technology.
- This indicates customers spend more on tech products.
- Business focus could be on expanding Technology category inventory.

PROFIT BY CATEGORY

```python
# Group profit by category
category_profit = df.groupby("Category")["Profit"].sum().reset_index()

# Plot
plt.figure(figsize=(8,5))

sns.barplot(
    x="Category",
    y="Profit",
    data=category_profit
)

plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")

plt.grid(alpha=0.3)

plt.savefig("../images/profit_by_category.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Business Insights

- Technology category generates the highest profit.
- Some categories may have lower or even negative profitability.
- High sales do not always mean high profit.
- Business should focus on optimizing low-profit categories.

Sales by Region

```python
# Group sales by region
region_sales = df.groupby("Region")["Sales"].sum().reset_index()

# Plot
plt.figure(figsize=(8,5))

sns.barplot(
    x="Region",
    y="Sales",
    data=region_sales
)

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.grid(alpha=0.3)

plt.savefig("../images/sales_by_region.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Business Insights

- Sales vary significantly across regions.
- Some regions contribute more to total revenue than others.
- The company may focus on improving underperforming regions.
- Regional performance is not uniform, indicating market imbalance.

Monthly Sales Trend

```python
# Ensure Order Date is datetime
df["Order Date"] = pd.to_datetime(df["Order Date"])

# Create month column
df["Month"] = df["Order Date"].dt.to_period("M")

# Group sales by month
monthly_sales = df.groupby("Month")["Sales"].sum().reset_index()

# Convert Month back to string for plotting
monthly_sales["Month"] = monthly_sales["Month"].astype(str)

# Plot
plt.figure(figsize=(12,5))

sns.lineplot(
    x="Month",
    y="Sales",
    data=monthly_sales,
    marker="o"
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)
plt.grid(alpha=0.3)

plt.savefig("../images/monthly_sales_trend.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Business Insights

- Sales fluctuate over different months.
- Certain months show higher demand, indicating seasonal trends.
- The company can plan inventory based on peak months.
- Understanding time trends helps improve business forecasting.

Discount vs Profit

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Discount",
    y="Profit",
    data=df
)

plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")

plt.grid(alpha=0.3)

plt.savefig("../images/discount_vs_profit.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Business Insights

- Higher discounts tend to reduce profit.
- Some heavily discounted orders result in losses.
- Discount strategy must be optimized carefully.
- Not all sales growth is profitable growth.

Correlation Heatmap

```python
# Select only numeric columns
numeric_df = df[["Sales", "Profit", "Discount", "Quantity"]]

# Correlation matrix
corr = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.savefig("../images/correlation_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Business Insights

- Sales and Profit show a positive relationship.
- Discount has a negative impact on Profit.
- Higher discounts often lead to reduced profitability.
- Quantity has a mild positive effect on sales.
- Pricing and discount strategy significantly influence profit.
